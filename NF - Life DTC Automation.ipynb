```python
import numpy as np
import pandas as pd
```

```python
file = r'''G:\Shared drives\Nationwide\2. (SHAREPOINT) Discipline Folders\4. Data + Tech\4. Business Intelligence\Reporting\2023 Delivered Reports + Process Docs\20230101 January Reports\DCM January.xlsx'''

clickthru_floodlights = ['LifeQuote_LandingPage_PageView',
                        'FreeWill_LandingPage_PageView',
                        'FreeWill_CreateWillNow_ButtonClick',
                        'FastApp_LandingPage_PageView',
                        'LifeInsuranceLearningCenter_LandingPage_PageView',
                        'InfoOrganizer_LandingPage_PageView',
                        'InfoOrganizer_GetYourCopy_ButtonClick']

conversion_floodlights = ['LifeQuote_GenderSelectionDropdown_Woman_ButtonClick',
                         'LifeQuote_GenderSelectionDropdown_Man_ButtonClick',
                         'LifeQuote_SeeYourTermLifeQuote_ButtonClick',
                         'FastApp_CompletedLandingPage_PageView',
                         'FastApp_GetYourQuote_ButtonClick',
                         'KPI - Life Insurance Schedule / Call Now']


campaign_name = 'NAI_NIF_Life DTC_2023'

def create_dcm_pivot(file,clickthru_floodlights,conversion_floodlights,campaign_name):  
    
    #dcm = pd.read_excel(file).iloc[9:]
   # dcm.columns=dcm.iloc[0]
    #dcm = dcm[1:]
    dcm = pd.read_excel(file)
    dcm['Prisma Code'] = dcm['Placement'].astype(str).str[:7]
    dcm = dcm.loc[dcm['Campaign']==campaign_name]
    #dcm = dcm.loc[dcm['Site (CM360)']=='DoubleClick Bid Manager (DBM)']
    pivot = pd.pivot_table(dcm,index=dcm[['Prisma Code','Site (CM360)','Placement']],values =dcm[['Impressions','Clicks','Video Completions']],aggfunc='sum')
    ct_conv_df = dcm.loc[dcm['Activity'].isin(clickthru_floodlights)]
    ct_conv_df = ct_conv_df[['Prisma Code','Site (CM360)','Placement','Activity','Click-through Conversions']]
    ct_conv_pivot = pd.pivot_table(ct_conv_df,index=ct_conv_df[['Prisma Code','Site (CM360)','Placement']],columns='Activity',values='Click-through Conversions',aggfunc='sum')
    merged_pivot = pd.merge(pivot,ct_conv_pivot, how='left',on=['Prisma Code','Site (CM360)','Placement']).fillna(0)
    tot_conv_df = dcm.loc[dcm['Activity'].isin(conversion_floodlights)]
    tot_conv_df = tot_conv_df[['Prisma Code','Site (CM360)','Placement','Activity','Total Conversions']]
    tot_conv_pivot = pd.pivot_table(tot_conv_df,index=tot_conv_df[['Prisma Code','Site (CM360)','Placement']],columns='Activity',values='Total Conversions',aggfunc='sum')    
    tot_merged_pivot = pd.merge(merged_pivot,tot_conv_pivot, how='left',on=['Prisma Code','Site (CM360)','Placement']).fillna(0)
    tot_merged_pivot = tot_merged_pivot.reset_index()
    
    return tot_merged_pivot

create_dcm_pivot(file,clickthru_floodlights,conversion_floodlights,campaign_name)
```

```python
file_two = r'''G:\Shared drives\Nationwide\2. (SHAREPOINT) Discipline Folders\4. Data + Tech\4. Business Intelligence\Reporting\2023 Delivered Reports + Process Docs\20230101 January Reports\DV360 January.xlsx'''
def create_dv360_table(file_two):
    
    dv360 = pd.read_excel(file_two)
    dv360 = dv360[['Creative','Revenue (Adv Currency)']]   
    dv360['Prisma Code'] = dv360['Creative'].astype(str).str[:7]
    dv360=dv360.drop_duplicates()
    
    return dv360.groupby('Prisma Code').sum()

create_dv360_table(file_two)
```

```python
dcm_final = create_dcm_pivot(file,clickthru_floodlights,conversion_floodlights,campaign_name)
dv360_totals = create_dv360_table(file_two)
```

```python
all_dbm = pd.merge(dcm_final,dv360_totals,on='Prisma Code',how='left')
    
all_dbm
```

```python
writer = pd.ExcelWriter(r'''G:\My Drive\Life DTC - Jan.xlsx''')
all_dbm.to_excel(writer)

writer.save()
```
